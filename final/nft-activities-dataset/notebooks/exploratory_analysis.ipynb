# Exploratory Data Analysis on NFT Activities

In this notebook, we will perform exploratory data analysis (EDA) on the NFT activities dataset. We will visualize the data, summarize statistics, and gain insights into the NFT market.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')
```

```python
# Load the dataset
data_path = '../data/processed/nft_activities.csv'
nft_data = pd.read_csv(data_path)

# Display the first few rows of the dataset
nft_data.head()
```

```python
# Summary statistics
nft_data.describe()
```

```python
# Visualize the distribution of prices
plt.figure(figsize=(10, 6))
sns.histplot(nft_data['price'], bins=30, kde=True)
plt.title('Distribution of NFT Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Analyze activities over time
nft_data['timestamp'] = pd.to_datetime(nft_data['timestamp'])
activities_over_time = nft_data.groupby(nft_data['timestamp'].dt.date).size()

plt.figure(figsize=(12, 6))
activities_over_time.plot()
plt.title('NFT Activities Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Activities')
plt.xticks(rotation=45)
plt.show()
```

## Conclusion

In this exploratory analysis, we have visualized the distribution of NFT prices and analyzed the activities over time. Further analysis can be conducted to explore relationships between different variables.
